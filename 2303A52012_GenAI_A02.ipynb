##  1. (1 ponto) Develop mathematical modeling for the given case study and find solution using Python
Consider a chocolate manufacturing company that produces only two
types of chocolate i.e. A and B. Both the chocolates require Milk
and Choco only. To manufacture each unit of A and B, the following quantitiesaare

required:

Each unit of A requires 1 unit of Milk and 3 units of Choco

Each unit of B requires 1 unit of Milk and 2 units of Choco

The company kitchen has a total of 5 units of Milk and 12 units of
Choco. On each sale, the company makes a profit of Rs 6 per ant A
sold and Rs 5 per unit B sold.

Now, the company wishes to maximize its profit. How many units of
A and B should it produce respectively?

WhatsApp Image 2025-01-10 at 14.31.45_3436b7bd.jpg

```python
from scipy.optimize import linprog
import numpy as np
import math as mp

obj = [-6,-5]
A=[[1,1],[3,2]]
B=[5,12]

bounds=[(0,float('inf')),(0,float('inf'))]

opt=linprog(c=obj,A_ub=A,b_ub=B,bounds=bounds)
chac=np.round(opt.x)
print(f"no of A={chac[0]} B={chac[1]}\n")
max_profit=np.abs(np.round(opt.fun))
print(f"therefore. the company can produce {chac[0]} of units \'A\' and {chac[1]} units of \'B\' to get the max profits of RS {max_profit} ")
```

##  2. (1 ponto) Develop mathematical modeling for the given case study and find solution using Python
A farmer has recently acquired a 100 hectares piece of land. He has
decided to grow Wheat and barley on that land. Due to the quality
of the sun and the region's excellent climate, the entire production of
Wheat and Barley can be sold. He wants to know how to plant each
variety in the 100 hectares, given the costs, net profits and labor
requirements according to the data shown below

Screenshot 2025-01-10 144511.png

The farmer has a budget of Rs. 7,00,000 and availability of 1,200
man-days during the planning horizon. Find the optimal solution and
the optimal value.


WhatsApp Image 2025-01-10 at 14.31.45_4434656b.jpg

```python
from scipy.optimize import linprog
import numpy as np

obj = [-50,-120]
A=[[7000,2000],[10,30]]
B=[700000,1200]

A_eq=[[1,1]]
B_eq=[100]

bounds=[(0,float('inf')),(0,float('inf'))]

opt=linprog(c = obj, A_ub = A, b_ub = B, A_eq = A_eq, b_eq = B_eq, bounds = bounds , method='simplex')
chac=np.round(opt.x)
print(f"Number of A (Wheat) = {chac[0]}, B (Barley) = {chac[1]}")
max_profit = -opt.fun
print(f"The farmer should plant 'Wheat' on {chac[0]} hectares and 'Barley' on {chac[1]} hectares to get the maximum profit of Rs {max_profit:.2f}.\n\n")
```
